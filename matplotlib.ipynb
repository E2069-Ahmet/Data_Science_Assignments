```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, label = "X Squared")
ax.plot(x,x**3, label = "X Cubed")
ax.legend(loc=1) # default 0, loc = 1'den 10'a kadar deer alabilir.
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, label = "X Squared")
ax.plot(x,x**3, label = "X Cubed")
ax.legend(loc=(0.1,0.1)) # default 0, loc = 1'den 10'a kadar deer alabilir.
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = "#FF8C00") #RGB Hex Code google bak
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = "green") #RGB Hex Code google bak, 16 kodlardr bunlar, daha farkl renkler elde etmek iin kullanlr
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = "purple", linewidth = 5) #linewidth =1 default
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = "purple", lw = 3, alpha =0.1) # alpha rengin saydamln ayarlyor
                                                            # linewidth lw olarak gsterilebilir.
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = "purple", lw = 3, linestyle = "-.") # linestyle ls yazlabiliyor, "-", "--", ":","-.", "steps"
```
